Task2

```python
l1 = [1, 2, 3, 4, 5]
l2 = [5, 6, 7, 8,9]
common_count = 0
for elem in l1:
    if elem in l2:
        common_count += 1
if common_count == 2:
    print("True")
else:
    print("False")
```

```python
lst = [1, 2, 3, 4]
def l2nd(lst):
    if len(lst) == 1:
        return {lst[0]: {}}
    else:
        return {lst[0]: l2nd(lst[1:])}
print(l2nd(lst))
```
