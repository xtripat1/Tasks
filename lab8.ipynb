```python
x= ['Madam', 'is', 'Madam']
def IsPalindrome(x):
  r_x =x[::-1]
  if x==r_x:
    return True
  else:
    return False


IsPalindrome(x)
```

```python
s1=list(input("Enter 1st string: "))
s2=list(input("Enter 2nd string: "))

def is_anagram(s1, s2):
  if sorted(s1) == sorted(s2):
    return True
  else:
    return False

is_anagram(s1,s2)
```

```python
a= int(input("the	number	of	a	word	in	the	Fibonacci	sequence : "))

def F_S(a):
  if a<0:
    return 0
  elif a==1:
    return 1
  else :
    return F_S(a-1) + F_S(a-2)

value= F_S(a)
print("The value of", a, "element is",value )
```
