Task 2

```python
x = input("Print a sentence with letters and numbers: ")
c_l = 0
c_n = 0
for i in x:
    if i.isalpha():
        c_l += 1
    elif i.isnumeric():
        c_n += 1
print("Letters:", c_l)
print("Numbers:", c_n)
```

Task 3

```python
text = input("Enter some text: ")
counts = {}

for char in text:
    if char.isalpha():
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1

for char, count in sorted(counts.items()):
    print(char + ": ",count)
```

```python
def is_pw_valid(pw):
    lowercase = False
    uppercase = False
    digit = False
    special = False

    if len(pw) < 6 or len(pw) > 12:
        return False

    for char in pw:
        if char.islower():
            lowercase = True
        elif char.isupper():
            uppercase = True
        elif char.isdigit():
            digit = True
        elif char in ['$','#','@']:
            special = True

    if lowercase and uppercase and digit and special:
        return True
    else:
        return False

pw = input("Enter a password: ")
if is_pw_valid(pw):
    print("Password is valid")
else:
    print("Password is not valid")
```
