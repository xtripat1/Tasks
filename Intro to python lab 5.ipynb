```python
list1 = [2, 58, 4, 12, 6, 8, 7, 9]
list2 = [12, 9, 25, 8, 64, 58]

list3=[]
for i in list1:
  if i not in list2:
    list3.append(i)
for j in list2:
  if j not in list1:
    list3.append(j)

print(list3)
```
