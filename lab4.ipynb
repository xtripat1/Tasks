```python
sum = 0
neg_count = 0
prev_num = None

while True:
    num = int(input("Enter a number: "))
    if sum + num > 100:
        print("Sum of numbers exceeds 100")
        break

    if num < 0:
        neg_count += 1
        if neg_count > 10:
            print("More than 10 negative numbers entered")
            break

    if prev_num == num:
        print("Two consecutive numbers have the same value")
        break
 
    sum += num
    prev_num = num

print("End")
```

```python
alcohol_ml = float(input("Enter the amount of alcohol consumed in milliliters: "))
alcohol_percent = float(input("Enter the percentage of alcohol in the drink: "))
weight = float(input("Enter your weight in kilograms: "))
gender = input("Enter your gender (male or female): ")
alcohol_amount = alcohol_ml * alcohol_percent / 100 * 0.8 # 0.8 is the density of alcohol in g/ml
k = 0.7 if gender == 'male' else 0.6
bac = alcohol_amount / (k * weight)
print("Your blood alcohol content is {:.2f} per mille".format(bac))
```
